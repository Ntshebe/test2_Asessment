```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/Admin/Downloads/bky6-3za5.csv")

#for easy to sea my code is running

raws = 5
column_Name = 'pocket_name'
data_column = df[column_Name]
data_raw = data_column.head(raws)

column = 'density_dwellings'
Name = df[column]
data = Name.head(raws)


plt.pie(data, labels=data_raw,autopct='%1.1f%%',startangle=140)
plt.title('POCKET NAME AND DENSITY DWELLINGS')
plt.show()
```

```python
selector = df.iloc[0:5]

plt.figure(figsize=(10, 6))
sns.countplot(data=selector, x='structure_count')
plt.title('Count of Informal Settlements by Structure count')
plt.xlabel('Structure count')
plt.ylabel('Count')
plt.show()
```

```python
selecter= df.iloc[0:5]
sns.pairplot(selecter[['avg_hhsize', 'structure_count', 'pocket_code']])
plt.title('Structure_count, pocket_code and avg_hhsize')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:/Users/Admin/Downloads/bky6-3za5.csv")

#for easy to sea my code is running

select = df.iloc[0:5]

plt.figure(figsize= (10,5))

#bar plot

plt.bar(select.pocket_name,select.density_dwellings, color='blue' ,alpha=0.7)
plt.title('POCKET NAME AND DENSITY DWELLINGS ')
plt.ylabel('Density Dwelling')
plt.xlabel('Pocket Name')
plt.grid(True)
plt.show()
```

```python
selected = df.iloc[0:5]
plt.figure(figsize=(10, 6))
sns.boxplot(data=selected, x='density_dwellings', y='structure_count')
plt.title('Distribution of Structure by density dwellings')
plt.xlabel('density Dwelling')
plt.ylabel('Structure Count')
plt.show()
```
